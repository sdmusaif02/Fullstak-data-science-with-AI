```python
data = {
    "salary" : [25000, 30000, 37000, 39000, 48000, 55000, 35000, 27000, 31000, 76000, 80000, 21000, 52000, 53000, 87000, 65000, 75000, 85000, 95000]
                }
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.shape
```

```python
#line plot:
plt.plot(df["salary"], color = "green", marker = "o" ,linestyle = ":", linewidth = "2")
plt.grid()
plt.show()
```

```python
#histogram
plt.hist(df["salary"], bins =5, color ="green", edgecolor = "black")
plt.show()
```

```python
plt.boxplot(df["salary"])
plt.show()
```

```python
df.loc[20] = [0]
plt.boxplot(df["salary"])
plt.show()
```

```python
#Univariante: categorical
```

```python
df["dept"] = ['HR', 'IT', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR','IT','HR']*2
```

```python
df.head()
```

```python
# pie chart:
count = df["dept"].value_counts()
count
```

```python
plt.pie(count, labels=count.index)
plt.show()
```

```python
plt.pie(count, labels=count.index, autopct="%1.0f%%")
plt.show()
```

```python
plt.pie(count, labels=count.index, autopct="%1.0f%%",  explode = [0, 0.1, 0])
plt.show()
```

```python
plt.pie(count, labels=count.index, autopct="%1.0f%%",  explode = [0, 0.1, 0.2])
plt.axis("equal")
plt.show()
```

```python
#countplot
plt.bar(count.index,count,edgecolor = "black")
plt.show()
```

```python
plt.bar(count.index,count, color = ["green", "orange","yellow"],edgecolor = "black")
plt.show()
```

```python
df["Age"] = [22, 25, 29, 24, 30, 35, 40, 36, 28, 22, 23, 25, 26, 24, 31, 29, 27, 33, 35, 47,]
```

```python
df.head()
```

```python
#scatter plot:
```

```python
plt.scatter(df["Age"],df["salary"],color = "orange")
plt.show()
```

```python
plt.plot(df["Age"], df["salary"], color = "green", marker = "o", linestyle = ":", linewidth = "2")
plt.grid()
plt.show()
```

```python
# Sort the DataFrame by Age
sort_age = df.sort_values("Age")

# Plot Age vs Salary using the sorted DataFrame
plt.plot(sort_age["Age"], sort_age["salary"], color="red", marker="o", linestyle=":", linewidth=2)
plt.grid()
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")
plt.show()
```

```python
# Bar chart:
```

```python
plt.bar(sort_age["Age"],df["salary"],color = "purple", edgecolor = "black", align="edge")
plt.show()
```
