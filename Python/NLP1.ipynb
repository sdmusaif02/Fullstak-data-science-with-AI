```python
# Function: POS tag ko WordNet format me convert kare
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # default

lemmatizer = WordNetLemmatizer()

text = "The cats were running faster than the dogs."
tokens = nltk.word_tokenize(text)
pos_tags = nltk.pos_tag(tokens)

lemmatized_words = [
    lemmatizer.lemmatize(word, get_wordnet_pos(pos))
    for word, pos in pos_tags
]

print("Original:", tokens)
print("Lemmatized:", lemmatized_words)
```

```python
stop_words = set(stopwords.words('english'))

filtered_words = [w for w in lemmatized_words if w.lower() not in stop_words and w.isalpha()]

print("After Stopword Removal:", filtered_words)
```
